```python
compras = compras.drop(45)
```

```python
compras.drop(34, inplace=True)
```

```python
compras[compras.isnull().any(axis=1)]
```

```python
compras.loc[33, "Total"] = compras.loc[33, "Unidade"] * \
compras.loc[33, "Preço Unidade"]
```

```python
compras.loc[40, "Total"] = compras.loc[40, "Unidade"] * \
compras.loc[40, "Preço Unidade"]
```

```python
compras.loc[42, "Preço Unidade"] = \
compras.loc[42, "Total"]/compras.loc[42, "Unidade"]
```

```python
compras
```

```python
compras[compras.isnull().any(axis=1)]
```

A base tem formatação diferente?

```python
compras.head(10)
```

```python
compras.Data = compras.Data.str.replace("-", "/")
```

```python
compras.Data = compras.Data.str.replace("//", "/")
```

```python
compras
```

```python
preco_ajustado = []
for preco in compras["Preço Unidade"]:
    preco_ajustado.append(float(str(preco).replace(";", ".")))
```

```python
compras["Preço Unidade"] = preco_ajustado
```

```python
compras.to_excel("compras_ajustado.xlsx")
```
